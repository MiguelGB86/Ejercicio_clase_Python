### Validamos Nulos por fila y columna para los 2 Dataframes creados

```python
# Validamos NA´s por Fila y Columna para los 2 Dataframes

print("\n")
def num_nulos(x):
  return x.isnull().sum()

print("Número de NA's por fila Dataframe Test: ")
print(df.apply(num_nulos, axis=1))

print("\n")
print("Número de NA's por columna Dataframe Test: ")
print(df.apply(num_nulos, axis=0))

print("\n")
print("Número de NA's por fila Dataframe Train: ")
print(df1.apply(num_nulos, axis=1))

print("\n")
print("Número de NA's por columna Dataframe Train: ")
print(df1.apply(num_nulos, axis=0))
```

### Generamos Boxplot para análisis de datos, considerando el dataframe Train.

#### Cantidad de Préstamo por Historial Crediticio. 

```python
#Boxplot Cantidad de Préstamo por Historial Crediticio

print("\n")
print("Gráfico Boxplot Cantidad de Préstamo por Historial Crediticio")
df1.boxplot(column='LoanAmount', by='Credit_History',figsize=(15,8))
```

#### Lo que observamos en el gráfico Boxplot es lo siguiente: 

###### - Si se tiene Historial Crediticio el valor minímo es menor con respecto a cuando no se tiene Historial. En el gráfico se puede observar que es más o menos de 40 el valor mínimo cuando no se tiene historial. 
###### - El primer cuartil para historial crediticio es de 100, mientras que para los que no lo tienen varia ligeramente.
###### - La mediana es muy similar tanto para los que tienen como no tienen historial crediticio. Se puede observar que el valor oscila en 120 aproximadamente.  
###### - El tercer cuartil es muy similar en ambos casos y practicamente oscila en un valor de 160. 
###### - En el valor máximo también se nota similitud en ambos casos. 
###### - Existen valores atipicos; sin embargo ambos casos se encuentran en un máximo que oscila en 600, cuando se tiene historial crediticio se observan mayor número de casos atipicos. 

###### - Dado lo anterior lo que se puede concluir es que se puede dar mayor número de casos atipicos cuando se cuenta con Historial Crediticio. 

#### Cantidad de Préstamo por Género. 

```python
#Boxplot Cantidad de Préstamo por Género

print("\n")
print("Gráfico Boxplot Cantidad de Préstamo por Historial Crediticio")
df1.boxplot(column='LoanAmount', by='Gender',figsize=(15,8))
```

#### Lo que observamos en el gráfico Boxplot es lo siguiente:

###### - En la mediana igualmente se observa diferencia.
###### - En el tercer cuartil se observa una diferencia considerable entre el género masculino y femenino, en las mujeres se observan valores por debajo de 150; mientras que en los hombres es por arriba.
###### - En el valor máximo se nota una gran diferencia, las mujeres un poco arriba de 200; mientras que para los hombres se observa que casi llega a 300.
###### - Existen mayor número de casos atipicos en el género masculino.

###### - Dado lo anterior lo que se puede concluir es que existe diferencia en las cantidades de préstamo dependiendo el sexo, teniendo mayor oportunidad y monto de préstamo el género masculino.
